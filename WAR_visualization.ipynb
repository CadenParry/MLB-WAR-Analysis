```python
import pandas as pd 

contracts = pd.read_csv('data/hitter_contracts.csv', sep=',')
player_stats = pd.read_csv('website/player_visualization.csv', sep=',')
player_stats.head()

player_stats = player_stats.dropna()
contracts = contracts.dropna()


```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter, MaxNLocator
import numpy as np


plt.xlabel('Year')
plt.ylabel('WAR')

plt.ylim([-3, 13])

# amake the color of the opaque





plt.legend(loc='lower left')

selected_player = player_stats[player_stats['player_id'] == 'rizzoan01']

selected_player
player_year = selected_player['yearID']
player_war = selected_player['WAR']

plt.plot(player_year, player_war)
plt.scatter(player_year, player_war, label='Not in Largest Contract')


# Set xticks every 2 years
start_year = player_year.min()
end_year = player_year.max() + 1
plt.xticks(np.arange(start_year, end_year, 2))

largest_span = selected_player[~selected_player['out_of_contract']]
player_year = largest_span['yearID']
player_war = largest_span['WAR']

player_war

plt.plot(player_year, player_war, color='#FFA500')
plt.scatter(player_year, player_war, label='In Largest Contract', color='#FFA500')

plt.title('Barry Bonds WAR Over Time')
# Add a legend in the bottom left


plt.axhline(y=8, color='green', alpha = 0.5, linestyle='--', label = 'MVP Candidate')
plt.axhline(y=5, color='blue', alpha = 0.5, linestyle='--', label = 'All Star')
plt.axhline(y=2, color='orange', alpha = 0.5, linestyle='--', label='Substitue Player')
plt.axhline(y=0, color='red', alpha = 0.5, linestyle='--', label='Replacement Level')

plt.legend(loc='lower left')


# Set xticks to every year in player_y
plt.show()
```
