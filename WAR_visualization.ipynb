```python
import pandas as pd 

contracts = pd.read_csv('data/hitter_contracts.csv', sep=',')
player_stats = pd.read_csv('website/player_visualization.csv', sep=',')
player_stats.head()

contracts = contracts.dropna()

```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter, MaxNLocator
import numpy as np


plt.xlabel('Year')
plt.ylabel('WAR')

plt.ylim([-3, 13])

plt.axhline(y=0, color='red', linestyle='--')
plt.axhline(y=2, color='orange', linestyle='--')
plt.axhline(y=5, color='blue', linestyle='--')
plt.axhline(y=8, color='green', linestyle='--')

selected_player = player_stats[player_stats['player_id'] == 'bondsba01']


selected_player
player_year = selected_player['yearID']
player_war = selected_player['WAR']



plt.plot(player_year, player_war, label='WAR')
plt.scatter(player_year, player_war, label='WAR')


# Set xticks every 2 years
start_year = player_year.min()
end_year = player_year.max() + 1
plt.xticks(np.arange(start_year, end_year, 2))

largest_span = selected_player[~selected_player['out_of_contract']]
player_year = largest_span['yearID']
player_war = largest_span['WAR']

player_war

plt.plot(player_year, player_war, label='WAR', color='#FFA500')
plt.scatter(player_year, player_war, label='WAR', color='#FFA500')



# Set xticks to every year in player_y
plt.show()
```
