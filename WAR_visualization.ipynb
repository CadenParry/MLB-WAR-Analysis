```python
import pandas as pd 

contracts = pd.read_csv('data/hitter_contracts.csv', sep=',')
player_stats = pd.read_csv('data/player_visualization.csv', sep=',')
player_stats.head()

```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter, MaxNLocator
import numpy as np


plt.xlabel('Year')
plt.ylabel('WAR')

plt.ylim([-3, 12])

plt.axhline(y=0, color='red', linestyle='--')
plt.axhline(y=2, color='orange', linestyle='--')
plt.axhline(y=5, color='magenta', linestyle='--')
plt.axhline(y=8, color='green', linestyle='--')

slected_player = player_stats[player_stats['player_id'] == 'griffke02']
player_year = slected_player['yearID']
player_war = slected_player['WAR']

plt.plot(player_year, player_war, label='WAR')


# Set xticks every 2 years
start_year = player_year.min()
end_year = player_year.max() + 1
plt.xticks(np.arange(start_year, end_year, 2))

largest_span = slected_player[~slected_player['out_of_contract']]
player_year = largest_span['yearID']
player_war = largest_span['WAR']

plt.plot(player_year, player_war, label='WAR', color='#FFA500')

print(plt.colormaps()[5])

# Set xticks to every year in player_y
plt.show()
```
